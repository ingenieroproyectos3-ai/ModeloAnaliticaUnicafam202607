## **AGREGAR UN DATO NUEVO**

```python
# Diccionario con los datos personales

perfil={
    "Nombre":"Ivan Reina",
    "Edad":"30",
    "Ciudad":"Bogotá",
    "Hobbies":["Fútbol","Viajar","Videojuegos"]
}

# Menú

print("---MENÚ DE MI PERFIL---")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato")
print("3. Eliminar un dato")
print("4. Consultar un dato")
print("5. Mostrar el perfil")
print("6. Salir")

opcion=input("Seleccione una opción: ")

# Agregar un dato nuevo
if opcion=="1":
  clave=input("Ingrese el nuevo dato personal: ")
  valor=input("Ingrese el valor de ese dato personal: ")
  perfil[clave]=valor
  print("\n Perfil actualizado: ")
  print(perfil)

# Actualizar un dato existente
elif opcion == "2":
    clave = input("Ingrese el dato personal que desea actualizar: ")

    if clave in perfil:
        nuevo_valor = input("Ingrese el nuevo valor del dato: ")
        perfil[clave] = nuevo_valor
        print("\nDato actualizado correctamente.")
        print(perfil)
    else:
        print("El dato personal no existe en el diccionario.")

# Eliminar un dato con "del"
elif opcion == "3":
    clave = input("Ingrese el dato personal que desea eliminar: ")
    if clave in perfil:
        del perfil[clave]
        print("\nDato personal eliminado correctamente.")
        print(perfil)
    else:
        print("El dato personal no existe en el diccionario.")

# Eliminar un dato con ".pop()"
elif opcion=="4":
      clave = input("Ingrese el dato personal que desea eliminar: ")
      if clave in perfil:
          valor = perfil.pop(clave)
          print("\nDato personal eliminado correctamente.")
          print(perfil)
      else:
          print("La clave no existe en el diccionario.")

# Consultar un dato personal
elif opcion == "5":
    clave = input("Ingrese el dato personal que desea consultar: ")
    valor = perfil.get(clave, "No disponible")
    print(clave, valor)

# Mostrar el perfil completo
elif opcion == "6":
    print("\n ---Mi perfil completo:")
    print(perfil)
else:
  print("Opción no válida.")
```

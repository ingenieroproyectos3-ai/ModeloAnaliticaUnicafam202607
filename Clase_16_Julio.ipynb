## Ejercicio 1

```python
#Escribe un programa que sume todos los numeros pares del 1 al 100 e imprima el resultado
suma = 0

for numero in range(1, 101):
    if numero % 2 == 0:
        suma+=numero

print("La suma de los números pares del 1 al 100 es:", suma)
```

## Ejercicio 2

```python
# Adivina el numero
num_sec=8
num=int(input("Ingrese un número: "))

while num !=num_sec:
  num=int(input("Ingrese un número nuevamente: "))
print("Felicitaciones adivinaste el número")
```

## Ejercicio 3

```python
num=int(input("Ingrese un número: "))
while num!=0:
  for i in range(1,11):
    mult=i*num
    print(f"{num}*{i}={mult}")
  num=int(input("Ingrese otro número: "))
print("Programa para tabla de multiplicar terminado")
```

## **FUNCIONES**

```python
# Crea una función que te salude de acuerdo a tu nombre
def saludar (nombre):
  print(f"Hola, esta es mi primer función, mi nombre es {nombre}")
saludar("Ivan")
```

```python
# Crea una función de Suma
def suma (a,b):
  return a+b
resultado=suma(2,2)
print(resultado)
```

```python
# Presentación con edad
def presentacion (nombre, edad=30):
  print(f"Hola mi nombre es {nombre} y tengo {edad} años")
presentacion("Ivan")
```

```python
# Genera una calculadora de operaciones
def operaciones (a,b):
  suma=a+b
  resta=a-b
  mult=a*b
  division=a/b
  return suma,resta,mult,division
result_suma, result_resta, result_mult, result_division= operaciones(8,5)
print(f"Suma: {result_suma}, Resta: {result_resta}, Multiplicación: {result_mult}, División: {result_division}")
```

```python
# Escriba una funcion que reciba la edad de una persona como parametro y determine si la persona es mayor de edad o menor de edad

def validar (edad_usuario):
    if edad_usuario>=18 :
      print(f"Tienes {edad_usuario} años, eres mayor de edad")
    else:
      print(f"Tienes {edad_usuario} años, eres menor de edad")
edad_usuario=int(input("Escriba la edad:"))
validar(edad_usuario)
```

```python
def clasificacion (nota_usuario):
    if nota_usuario>=0 and nota_usuario<50:
      print(f"Tu nota es {nota_usuario}. Clasificación: Insuficiente")
    elif nota_usuario>=50 and nota_usuario<70:
      print(f"Tu nota es {nota_usuario}. Clasificación: Suficiente")
    elif nota_usuario>=70 and nota_usuario<90:
      print(f"Tu nota es {nota_usuario}. Clasificación: Bueno")
    elif nota_usuario>=90 and nota_usuario<=100:
      print(f"Tu nota es {nota_usuario}. Clasificación: Excelente")
    else:
      print("La nota ingresada es inválida")
nota_usuario=int(input("Ingrese la nota para clasificar:"))
clasificacion(nota_usuario)
```

```python
def check(new_password, password="secreta123"):
    if new_password == password:
        print("La contraseña es correcta")
    else:
        print("La contraseña es incorrecta")

new_password = input("Ingrese la contraseña: ")
check(new_password)
```

```python
def clasificacion (imc):

    if imc<18.5:
      print(f"Tu IMC es {imc}. El peso de paciente se clasifica como: Bajo peso")
    elif imc>=18.5 and imc<25:
      print(f"Tu IMC es {imc}. El peso de paciente se clasifica como: Normal")
    elif imc>=25 and imc<30:
      print(f"Tu IMC es {imc}. El peso de paciente se clasifica como: Sobrepeso")
    else:
      print(f"Tu IMC es {imc}. El peso de paciente se clasifica como: Obesidad")

peso=float(input("Ingrese su peso en kg:"))
altura=float(input("Ingrese su altura en m:"))
imc=float (peso/(altura**2))
clasificacion (imc)
```

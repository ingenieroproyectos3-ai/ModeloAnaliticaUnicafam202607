```python
#3. Dada precios = [12000, 45000, 8500, 67000, 3200], usa un ciclo para: (1) imprimir cada precio con formato Item 1: $12,000, (2) imprimir al final cuántos artículos cuestan más de $10,000.
precios=[12000, 45000, 8500, 67000, 3200]
costoso=0

for i, x in enumerate(precios,1):
    print(f"Item {i}: ${x:,}")
    if x > 10000:
        costoso += 1
print(f"Hay {costoso} artículos que cuestan más de $10,000")
```

```python
#4. Un banco tiene saldo = 1000. Cada mes retiras 150 y se cobran 20 de comisión. Usa un ciclo while para simular cuántos meses pasan antes de que el saldo llegue a cero o menos. Imprime el saldo al final de cada mes.
saldo = 1000
mes   = 0
while saldo > 0:
    saldo -= 150
    saldo -= 20
    mes   += 1
    print(f"Mes {mes}: Saldo ${saldo}")
print(f"Su salfo se agota en {mes} meses")
```

```python
#5. Dada la lista numeros = [3, 7, -2, 15, 4, -8, 9, 1], escribe un ciclo que: (1) saltee los números negativos con continue, (2) se detenga con break al encontrar un número mayor a 10, (3) imprima los números que procesa.
numeros = [3, 7, -2, 15, 4, -8, 9, 1]
for n in numeros:
    if n < 0:
        continue
    if n > 10:
        print(f"Se encuentran {n}")
        break
    print(f"Procesando: {n}")
```
